##Imputing or Deleting values

```python
# Filling missing values (e.g., filling total_bedrooms with the median)
median_bedrooms = housing['total_bedrooms'].median()
median_bedrooms
```

```python
housing['total_bedrooms'].fillna(0,inplace=True)
```

```python
mean_value = housing['total_bedrooms'].mean()
housing['total_bedrooms'].fillna(mean_value, inplace=True)
```

```python
from scipy import stats

# Calculate Z-scores
z_scores = stats.zscore(housing.select_dtypes(include=[float, int]))

# Detect outliers (typically Z-score > 3)
outliers = (abs(z_scores) > 3).any(axis=1)

# Print rows with outliers
print(housing[outliers])
```

#Joining Data frames

```python
near_bay_homes = housing[housing['ocean_proximity'] == 'NEAR BAY']
other_homes = housing[housing['ocean_proximity'] != 'NEAR BAY']
```

```python
# Concatenating the DataFrames vertically
concatenated_df = pd.concat([near_bay_homes, other_homes], axis=0)
concatenated_df
```

```python
# Mapping old values to new ones
proximity_map = {
    'NEAR BAY': 'NBAY',
    'INLAND': 'INL',
    'NEAR OCEAN': 'NOCEAN',
    '<1H OCEAN': 'OCEAN1H',
    'ISLAND': 'ISL'
}

# Use map to apply the changes
housing['ocean_proximity_short'] = housing['ocean_proximity'].map(proximity_map)

# Show the result
print(housing[['ocean_proximity', 'ocean_proximity_short']].head())
```
